```python
import bisect

def generate_subset_sums(arr, idx, current_sum, result):
    if idx == len(arr):
        result.append(current_sum)
        return
    # Exclude current element
    generate_subset_sums(arr, idx + 1, current_sum, result)
    # Include current element
    generate_subset_sums(arr, idx + 1, current_sum + arr[idx], result)


def closest_subset_sum(arr, S):
    n = len(arr)
    mid = n // 2

    left = arr[:mid]
    right = arr[mid:]

    left_sums = []
    right_sums = []

    generate_subset_sums(left, 0, 0, left_sums)
    generate_subset_sums(right, 0, 0, right_sums)

    right_sums.sort()

    ans = float("inf")

    for l_sum in left_sums:
        target = S - l_sum
        idx = bisect.bisect_left(right_sums, target)

        if idx < len(right_sums) and right_sums[idx] == target:
            return 0  # exact match found
        if idx < len(right_sums):
            ans = min(ans, abs(l_sum + right_sums[idx] - S))
        if idx - 1 >= 0:
            ans = min(ans, abs(l_sum + right_sums[idx - 1] - S))

    return ans


# Driver code
T = int(input().strip())
for _ in range(T):
    N, S = map(int, input().split())
    arr = list(map(int, input().split()))
    print(closest_subset_sum(arr, S))
```
